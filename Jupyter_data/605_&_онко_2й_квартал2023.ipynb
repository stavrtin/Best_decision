```python
# 111 # -------------------------------------------------  Количество пациентов в реестре, чел. на date_start  ALL & ONKO (все  - взрос - дети)------
def def_data_count_start_year_onko(df, start, finish):
 
   
    data_count_start_year = df.loc[ 
                                          (df['Дата_вкл_реестр'] <= start) & ( (df['Дата_выбытия'].isnull())   | (df['Дата_выбытия'] >= start)   
                )].shape[0]
    
    data_count_start_year_onko_all = df.loc[ 
                                         ( (df['Дата_вкл_реестр'] <= start)   & ( (df['Дата_выбытия'].isnull())   | (df['Дата_выбытия'] >= start)  ) 
                                          
                                         & ( ~(df['Онкология'].isnull()))   
                )].shape[0]
   

    data_count_start_year_onko_vzros = df.loc[ 
                                         ( (df['Дата_вкл_реестр'] <= start)   & ( (df['Дата_выбытия'].isnull())   | (df['Дата_выбытия'] >= start)  ) 
                                          
                                         & ( ~(df['Онкология'].isnull()))   
                                         & ( (df['Дети'].isnull()))   
                )].shape[0]

    data_count_start_year_onko_deti = df.loc[ 
                                         ( (df['Дата_вкл_реестр'] <= start)   & ( (df['Дата_выбытия'].isnull())   | (df['Дата_выбытия'] >= start)  ) 
                                          
                                         & ( ~(df['Онкология'].isnull()))   
                                         & ( ~(df['Дети'].isnull()))   
                )].shape[0]


    return  data_count_start_year, data_count_start_year_onko_all, data_count_start_year_onko_vzros,   data_count_start_year_onko_deti
```

```python
# 222 # ------------------------------------------ --- Поступило (чел.) ALL & ОНКО (все  - взрос - дети) -----  ------
def def_data_new_v_period(df, start, finish):
    
    data_new_v_period_all = df.loc[ (df['Дата_вкл_реестр'] < finish) & (df['Дата_вкл_реестр'] >= start)].shape[0]
   
    data_new_v_period_onko = df.loc[ 
                                         ( (df['Дата_вкл_реестр'] >= start)   & ( (df['Дата_вкл_реестр'] < finish)  ) 
                                         & ( ~(df['Онкология'].isnull()))   
                )].shape[0]
  
    data_new_v_period_onko_vzros = df.loc[ 
                                         ( (df['Дата_вкл_реестр'] >= start)   & ( (df['Дата_вкл_реестр'] < finish)  ) 
                                          
                                         & ( ~(df['Онкология'].isnull()))   
                                         & ( (df['Дети'].isnull()))   
                )].shape[0]

    data_new_v_period_onko_deti = df.loc[ 
                                         ( (df['Дата_вкл_реестр'] >= start)   & ( (df['Дата_вкл_реестр'] < finish)  ) 
                                          
                                         & ( ~(df['Онкология'].isnull()))   
                                         & ( ~(df['Дети'].isnull()))   
                )].shape[0]

    return  data_new_v_period_all, data_new_v_period_onko, data_new_v_period_onko_vzros, data_new_v_period_onko_deti  
```

```python
# 33333 ----------------------------------------------  выбытие между .... Со статусом - "умер"
def def_data_dead(df, start, finish):

    data_dead = df.loc[ (df['Дата_выбытия'] <= finish) & (df['Дата_выбытия'] >= start) & (df['Текущий статус пациента'] == 'Умер') ].shape[0]

    return  data_dead
```

```python
# 4444 ---------------------------------------------  Количество пациентов, получателей помощи (чел.)
def def_data_all_clients(df, start, finish):

    data_all_clients = df.loc[ (df['Дата_вкл_реестр'] <= finish)  
                             &   ( (df['Дата_выбытия'].isnull())  | (df['Дата_выбытия'] >= start)     )].shape[0] 

    
    data_all_clients_onko = df.loc[ 
                                 (df['Дата_вкл_реестр'] <= finish)  &   ( (df['Дата_выбытия'].isnull())  | (df['Дата_выбытия'] >= start)     )
                                & ( ~(df['Онкология'].isnull()))   
        
                       ].shape[0] 

    data_all_clients_onko_vzros = df.loc[ 
                                 (df['Дата_вкл_реестр'] <= finish)  &   ( (df['Дата_выбытия'].isnull())  | (df['Дата_выбытия'] >= start)     )
                                & ( ~(df['Онкология'].isnull()))   
                                                 & ( (df['Дети'].isnull()))   
        
                       ].shape[0] 
    
    data_all_clients_onko_deti = df.loc[ 
                                 (df['Дата_вкл_реестр'] <= finish)  &   ( (df['Дата_выбытия'].isnull())  | (df['Дата_выбытия'] >= start)     )
                                & ( ~(df['Онкология'].isnull()))   
                                                 & ( ~(df['Дети'].isnull()))   
        
                       ].shape[0] 
    
    return  data_all_clients, data_all_clients_onko, data_all_clients_onko_vzros, data_all_clients_onko_deti
```

```python
def calculate_onko_v_period(df, date_start, date_finish):
    #--------------------------  Количество пациентов в реестре, чел.  на date_start  ------
    count_onko_v_period       = def_data_count_start_year_onko(df, start, finish)
    count_new              = def_data_new_v_period(df, start, finish)
    data_dead              = def_data_dead(df, start, finish)
    all_clients            = def_data_all_clients(df, start, finish)

   
    return count_onko_v_period,count_new, data_dead, all_clients
```
