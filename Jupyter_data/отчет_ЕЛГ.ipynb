```python
def calculate_data_elg(df, date_start, date_finish):
    # ---- было на дату старта в СТАЦИОНАРАХ ------
    data_new_year_elg = df_elg.loc[
                                   (df_elg['Дата_госпит'] <= date_start)
                               &   ( (df_elg['Дата_выбытия'].isnull())  | (df_elg['Дата_выбытия'] >= date_start)   
    )]
    
    data_new_year_elg = data_new_year_elg.drop_duplicates('Ключ')
    
    
    # ---- было на дату date_finish в СТАЦИОНАРАХ ------ 
    count_on_date_elg= df.loc[ 
                               (df['Дата_госпит'] <= date_finish)                                           
                           &   ((df['Дата_выбытия'].isnull()) | (df['Дата_выбытия'] >= date_finish))    
                           # &   (df['Категория'] == 'Старше трудосп')].shape[0]
                              ]
    count_on_date_elg = count_on_date_elg.drop_duplicates('Ключ')
    
    return data_new_year_elg.shape[0], count_on_date_elg.shape[0]
```
