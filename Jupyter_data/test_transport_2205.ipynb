```python
df_pik = pd.read_csv('тест_транспорт_пик_подг.csv', delimiter=';', encoding='cp1251')
df_pik.head(2)
```

```python
df_pik = df_pik[['Наименование филиала ', 'Необходимый тип транспортного средства',
       'Подана заявка', 'Выполнена заявка']]
```

```python
#  переименуем покороче столбцы
df_pik.rename(columns={'Наименование филиала ': 'филиал'}, inplace=True)
df_pik.rename(columns={'Необходимый тип транспортного средства': 'тип'}, inplace=True)
df_pik.rename(columns={'Подана заявка': 'Подана'}, inplace=True)
df_pik.rename(columns={'Выполнена заявка': 'Выполнена'}, inplace=True)
df_pik
```

```python
#  --------- DATATIME
df_pik['Выполнена'] = pd.to_datetime(df_pik['Выполнена'], dayfirst=True, errors = 'coerce').dt.normalize()
df_pik['Подана'] = pd.to_datetime(df_pik['Подана'], dayfirst=True, errors = 'coerce').dt.normalize()
df_pik
```

```python
# -------------------------------- работает по формированию НУЛЕВЫХ колонок ---------
def create_df_col(df, start, end):
    start_date = start
    end_date = end 
    df = df
    delta = datetime.timedelta(days=1)
    # data_list = []
    while (start_date <= end_date):
        
        df[start_date.strftime("%Y-%m-%d")] = 0  # создаем строку в DF
        start_date += delta
```
