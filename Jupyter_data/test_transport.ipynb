```python
# -------------------------------- работает по формированию НУЛЕВЫХ колонок ---------
def create_df_col(df, start, end):
    start_date = start
    end_date = end 
    df = df
    delta = datetime.timedelta(days=1)
    # data_list = []
    while (start_date <= end_date):
        
        df[start_date.strftime("%Y-%m-%d")] = 0  # создаем строку в DF
        start_date += delta
```
