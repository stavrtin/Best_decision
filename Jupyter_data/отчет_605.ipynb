```python
def calculate_data(df, date_start, date_finish):
    df_clr = df.copy()
    df_clr = df_clr.drop_duplicates()
    # ------  Количество пациентов в реестре, чел.  на date_start  ------
    data_count_start_year = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_start)         # дата внесения в реестр - ранее date_start
              & ( (df_clr['Дата_выбытия'].isnull())   | (df_clr['Дата_выбытия'] >= date_start)   # дата выбытия - еще не выбыл или дата выбытия - после date_start
                )].shape[0]
    
    # ---- Поступило (чел.)  -----
    data_count_all_period = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_finish) & (df_clr['Дата_первых_сведений'] >= date_start)].shape[0]
    # ---- из них Старше трудосп -------
    data_count_all_period_trud = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_finish) & (df_clr['Дата_первых_сведений'] >= date_start) & (df_clr['Категория'] == 'Старше трудосп')].shape[0]
    
    
    #  выбытие между .... Со статусом - "умер"
    data_dead = df_clr.loc[ (df_clr['Дата_выбытия'] < date_finish) & (df_clr['Дата_выбытия'] >= date_start) & (df_clr['Текущий статус пациента'] == 'Умер') ].shape[0]
    # ---- из них Старше трудосп -------
    data_dead_trud = df_clr.loc[ (df_clr['Дата_выбытия'] < date_finish) & (df_clr['Дата_выбытия'] >= date_start) & (df_clr['Текущий статус пациента'] == 'Умер')  & (df_clr['Категория'] == 'Старше трудосп')].shape[0]
    
    # Количество пациентов, получателей помощи (чел.) 
    data_all_clients = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_finish)  # дата внесения в реестр - ранее 01.01.22
                             &   ( (df_clr['Дата_выбытия'].isnull())  | (df_clr['Дата_выбытия'] >= date_start)                    )].shape[0]
    
    
    
    # --------дети -------------
    data_all_child = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_finish)  # дата внесения в реестр - ранее 01.01.22
                                     
         &   ( (df_clr['Дата_выбытия'].isnull())          # дата выбытия - еще не выбыл
              | (df_clr['Дата_выбытия'] >= date_start)   # или дата выбытия - после 01.01.22
           )
         & (df_clr['Дети'] == 'Дети')                     
      ].shape[0]
    
    # Количество пациентов, получателей помощи (чел.) --- Иногород ----------
    data_all_clients_inostranec = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_finish)  # дата внесения в реестр - ранее 01.01.22
                         &   ( (df_clr['Дата_выбытия'].isnull())  | (df_clr['Дата_выбытия'] >= date_start)) 
                             & (
                                 (df_clr['Иногород'] == 'Гражданин РФ/иногородний' ) 
                                   # | (df_clr['Иногород'] == 'Иностранец' ) 
                                  | (df_clr['Иногород'] == 'БОМЖ' ) 
                                   # |  (df_clr['Иногород'] == 'Украина' )
                                 |(df_clr['Иногород'] == 'ЛНР' ) 
                                 |(df_clr['Иногород'] == 'ДНР' ) )].shape[0]
    # ---- из них Старше трудосп -------
    data_all_clients_inostranec_trud = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_finish)  # дата внесения в реестр - ранее 01.01.22
                         &   ( (df_clr['Дата_выбытия'].isnull())  | (df_clr['Дата_выбытия'] >= date_start)) 
                             & (
                                 (df_clr['Иногород'] == 'Гражданин РФ/иногородний' ) 
                                   # | (df_clr['Иногород'] == 'Иностранец' ) 
                                  | (df_clr['Иногород'] == 'БОМЖ' ) 
                                   # |  (df_clr['Иногород'] == 'Украина' )
                                 |(df_clr['Иногород'] == 'ЛНР' ) 
                                 |(df_clr['Иногород'] == 'ДНР' ) )
                                             & (df_clr['Категория'] == 'Старше трудосп')].shape[0]

    # Количество пациентов, получателей помощи (чел.) --- Иногород-Дети ----------
    data_all_clients_inostranec_deti = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_finish)  # дата внесения в реестр - ранее 01.01.22
                         &   ( (df_clr['Дата_выбытия'].isnull())  | (df_clr['Дата_выбытия'] >= date_start)) 
                             & (
                                 (df_clr['Иногород'] == 'Гражданин РФ/иногородний' ) 
                                   # | (df_clr['Иногород'] == 'Иностранец' ) 
                                  | (df_clr['Иногород'] == 'БОМЖ' ) 
                                   # |  (df_clr['Иногород'] == 'Украина' )
                                 |(df_clr['Иногород'] == 'ЛНР' ) 
                                 |(df_clr['Иногород'] == 'ДНР' ) )
                                                 & (df_clr['Дети'] == 'Дети')].shape[0]

    return data_count_start_year, data_count_all_period, data_dead, data_all_clients, data_all_child, data_count_all_period_trud, data_dead_trud, data_all_clients_inostranec, data_all_clients_inostranec_trud, data_all_clients_inostranec_deti
```
