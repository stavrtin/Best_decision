```python
def self_apply(df, list_):
#     --------------------------------------------------самообращение --------------
    for i in list_:
        df.loc[     (df['Канал поступления информации'] == 'cамообращение')
                  |  (df['Кто направил'].str.contains(i) == True)              , 'Тип_канала'] = 'Самообращение'
    return df    
```

```python
def gp_apply(df, list_):
#     --------------------------------------------------ГП --------------
    for i in list_:
        df.loc[   (df['Кто направил'].str.contains(i) == True)                  , 'Тип_канала'] = 'ГП'
    return df    
```

```python
def caop_apply(df, list_):
#     --------------------------------------------------ЦАОП --------------
    for i in list_:
        df.loc[  (df['Кто направил'].str.contains(i) == True)                    , 'Тип_канала'] = 'ЦАОП'
    return df    
```

```python
def stac_apply(df, list_):
    # ------------------------------------------------------------------------- СТАЦИОНАРЫ 
    for i in list_:
        df.loc[
                   (df['Кто направил'].str.contains('цаоп') == False)
                 & (df['Кто направил'].str.contains(i) == True)              , 'Тип_канала'] = 'СТАЦИОНАРЫ'
    return df        
```

```python
def social_apply(df, list_):
    # ------------------------------------------------------------------------- Социальные 
    for i in list_:
        df.loc[df['Кто направил'].str.contains(i) == True                        , 'Тип_канала'] = 'Социальные'
    return df        
```
