```python
# 111 # -------------------------------------------------  Количество пациентов в реестре, чел.  на date_start  ------
def def_data_count_start_year(df, start, finish):
    
    df_clr = df
    date_start = start
    date_finish = finish
    data_count_start_year = df_clr.loc[ 
                                          (df_clr['Дата_первых_сведений'] <= date_start)         
                                      & ( (df_clr['Дата_выбытия'].isnull())   | (df_clr['Дата_выбытия'] >= date_start)   
                )].shape[0]
    data_count_start_year_trud = df_clr.loc[ (df_clr['Дата_первых_сведений'] <= date_start)         
                                      & ( (df_clr['Дата_выбытия'].isnull())   | (df_clr['Дата_выбытия'] >= date_start))  & (df_clr['Категория'] == 'Старше трудосп')].shape[0]
    
    return  data_count_start_year, data_count_start_year_trud  
```

```python
# 222 # ------------------------------------------ --- Поступило (чел.)  -----  ------
def def_data_count_all_period(df, start, finish):
    
    df_clr = df
    date_start = start
    date_finish = finish
    
    
    data_count_all_period = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_finish) & (df_clr['Дата_первых_сведений'] >= date_start)].shape[0]
    # ---- из них Старше трудосп -------
    data_count_all_period_trud = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_finish) & (df_clr['Дата_первых_сведений'] >= date_start) & (df_clr['Категория'] == 'Старше трудосп')].shape[0]

    return  data_count_all_period, data_count_all_period_trud  
```

```python
# 33333 ----------------------------------------------  выбытие между .... Со статусом - "умер"
def def_data_dead(df, start, finish):
    
    df_clr = df
    date_start = start
    date_finish = finish

    data_dead = df_clr.loc[ (df_clr['Дата_выбытия'] <= date_finish) & (df_clr['Дата_выбытия'] >= date_start) & (df_clr['Текущий статус пациента'] == 'Умер') ].shape[0]

    # ---- из них Старше трудосп -------
    data_dead_trud = df_clr.loc[ (df_clr['Дата_выбытия'] <= date_finish) & (df_clr['Дата_выбытия'] >= date_start) & (df_clr['Текущий статус пациента'] == 'Умер')  & (df_clr['Категория'] == 'Старше трудосп')].shape[0]

    return  data_dead, data_dead_trud  
```

```python
# 4444 ---------------------------------------------  Количество пациентов, получателей помощи (чел.)
def def_data_all_clients(df, start, finish):
    
    df_clr = df
    date_start = start
    date_finish = finish

    data_all_clients = df_clr.loc[ (df_clr['Дата_первых_сведений'] <= date_finish)  
                             &   ( (df_clr['Дата_выбытия'].isnull())  | (df_clr['Дата_выбытия'] >= date_start)     )].shape[0] 


    data_all_clients_trud = df_clr.loc[ (df_clr['Дата_первых_сведений'] <= date_finish) 
                             &   ( (df_clr['Дата_выбытия'].isnull())  | (df_clr['Дата_выбытия'] >= date_start))  
                             &   (  df_clr['Категория'] == 'Старше трудосп')                                       ].shape[0]
    
    return  data_all_clients, data_all_clients_trud
```

```python
# 5555 ------------------------------------------------дети -------------
def def_data_all_child(df, start, finish):
    
    df_clr = df
    date_start = start
    date_finish = finish

    data_all_child = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_finish)  
                             &   ( (df_clr['Дата_выбытия'].isnull())  | (df_clr['Дата_выбытия'] >= date_start   ))
                             &   (df_clr['Дети'] == 'Дети')                                                        ].shape[0]
    return  data_all_child
```

```python
# /6666  ----------------------------------------- Иногород ----------
def def_data_all_clients_inostranec(df, start, finish):
    
    df_clr = df
    date_start = start
    date_finish = finish
    
    data_all_clients_inostranec = df_clr.loc[ (df_clr['Дата_первых_сведений'] < date_finish)  # дата внесения в реестр - ранее 01.01.22
                         &   ( (df_clr['Дата_выбытия'].isnull())  | (df_clr['Дата_выбытия'] >= date_start)) 
                             & (
                                 (df_clr['Иногород'] == 'Гражданин РФ/иногородний' ) 
                               | (df_clr['Иногород'] == 'БОМЖ' ) 
                               |(df_clr['Иногород'] == 'ЛНР' )                  
                               # | (df_clr['Иногород'] == 'Иностранец' ) 
                               # |  (df_clr['Иногород'] == 'Украина' )
                               |(df_clr['Иногород'] == 'ДНР' ) )].shape[0]
        # ---- из них Старше трудосп -------
    
    data_all_clients_inostranec_trud = df_clr.loc[ 
                             (df_clr['Дата_первых_сведений'] < date_finish)  # дата внесения в реестр - ранее 01.01.22
                           & ( (df_clr['Дата_выбытия'].isnull())  | (df_clr['Дата_выбытия'] >= date_start)) 
                           & ( 
                               (df_clr['Иногород'] == 'Гражданин РФ/иногородний' ) 
                             | (df_clr['Иногород'] == 'БОМЖ' ) 
                              # | (df_clr['Иногород'] == 'Иностранец' ) 
                              # |  (df_clr['Иногород'] == 'Украина' )
                             |(df_clr['Иногород'] == 'ЛНР' ) 
                             |(df_clr['Иногород'] == 'ДНР' ) )
                            & (df_clr['Категория'] == 'Старше трудосп')].shape[0]
    
    return  data_all_clients_inostranec, data_all_clients_inostranec_trud    
```

```python
def calculate_data(df, date_start, date_finish):
    #--------------------------  Количество пациентов в реестре, чел.  на date_start  ------
    count_start_year       = def_data_count_start_year(df, start, finish)
    count_all_period       = def_data_count_all_period(df, start, finish)
    data_dead              = def_data_dead(df, start, finish)
    all_clients            = def_data_all_clients(df, start, finish)
    all_child              = def_data_all_child(df, start, finish)
    all_clients_inostranec = def_data_all_clients_inostranec(df, start, finish)

    
    return count_start_year, count_all_period, data_dead, all_clients, all_child, all_clients_inostranec
```
