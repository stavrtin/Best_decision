<b>  создадим функцию пробегания по датафрейму ПЕРЕВЕДЕННЫХ пациентов и исключением лишних строк
    </div>

```python
def df_transform(df_trans):
    df = df_trans.copy()
    df = df.sort_values('Дата_госпит')
    df['flag'] = 0
    for i, row_i in df.iterrows():
        for j, row_j in df.iterrows():
            if (row_i.Ключ == row_j.Ключ) and (row_j.Дата_госпит == row_i.Дата_выбыт) and (row_j.flag == 0):
                df.loc[i, 'Дата_выбыт'] = row_j.Дата_выбыт
                df.loc[i, 'Причина_выбыт'] = row_j.Причина_выбыт
                df.loc[i, 'Место_выбыт'] = row_j.Место_выбыт
                df.loc[i, 'Филиал'] = row_j.Филиал
                
                row_i.Дата_выбыт = row_j.Дата_выбыт
                row_i.Причина_выбыт = row_j.Причина_выбыт
                row_i.Место_выбыт = row_j.Место_выбыт
                row_i.Филиал = row_j.Филиал
                
                # df.drop(index= j, inplace=True)
                row_j.flag = 1
                df.loc[j, 'flag'] = 1
    df = df.loc[df['flag'] != 1]
    return df   
```

```python
def df_diagnost(df_diagn):
    df = df_diagn.copy()
    # df = df.sort_values('Дата_госпит')
    df['flag'] = 0
    df['Откуда_прибыл'].str.lower()
    df['Место_выбыт'].str.lower()
    

    df['Дата_госпит'] = pd.to_datetime(df['Дата_госпит'], dayfirst=True, errors = 'coerce').dt.normalize()
    df['Дата_выбыт'] = pd.to_datetime(df['Дата_выбыт'], dayfirst=True, errors = 'coerce').dt.normalize()

    
    
    for i, row_i in df.iterrows():
        for j, row_j in df.iterrows():
            
            if (row_i.Ключ == row_j.Ключ) and (row_j.gkb_status == row_i.gkb_perevod) and ((row_j.Дата_госпит - row_i.Дата_выбыт) <= pd.to_timedelta(14, unit='D') and ((row_j.Дата_госпит - row_i.Дата_выбыт) >= pd.to_timedelta(0, unit='D')) ) and (row_j.flag == 0):
                df.loc[i, 'Дата_выбыт'] = row_j.Дата_выбыт
                df.loc[i, 'Причина_выбыт'] = row_j.Причина_выбыт
                df.loc[i, 'Место_выбыт'] = row_j.Место_выбыт
                df.loc[i, 'Филиал'] = row_j.Филиал
                df.loc[i, 'diagnost'] = df.loc[i, 'diagnost'] + row_i.Место_выбыт
                                
                row_i.Дата_выбыт = row_j.Дата_выбыт
                row_i.Причина_выбыт = row_j.Причина_выбыт
                row_i.Место_выбыт = row_j.Место_выбыт
                row_i.Филиал = row_j.Филиал
                row_i.diagnost = row_i.diagnost + str(row_i.Место_выбыт)
                                
                row_j.flag = 1
                df.loc[j, 'flag'] = 1
                
    df = df.loc[df['flag'] != 1]
    return df   
```
