<b>  создадим функцию пробегания по датафрейму ПЕРЕВЕДЕННЫХ пациентов и исключением лишних строк
    </div>

```python
def df_transform(df_trans):
    df = df_trans.copy()
    df = df.sort_values('Дата_госпит')
    df['flag'] = 0
    for i, row_i in df.iterrows():
        for j, row_j in df.iterrows():
            if (row_i.Ключ == row_j.Ключ) and (row_j.Дата_госпит == row_i.Дата_выбыт) and (row_j.flag == 0):
                df.loc[i, 'Дата_выбыт'] = row_j.Дата_выбыт
                df.loc[i, 'Причина_выбыт'] = row_j.Причина_выбыт
                df.loc[i, 'Место_выбыт'] = row_j.Место_выбыт
                df.loc[i, 'Филиал'] = row_j.Филиал
                
                row_i.Дата_выбыт = row_j.Дата_выбыт
                row_i.Причина_выбыт = row_j.Причина_выбыт
                row_i.Место_выбыт = row_j.Место_выбыт
                row_i.Филиал = row_j.Филиал
                
                # df.drop(index= j, inplace=True)
                row_j.flag = 1
                df.loc[j, 'flag'] = 1
    df = df.loc[df['flag'] != 1]
    return df   
```

```python
def calculate_data_elg(df, date_start, date_finish):
    # ---- было на дату старта в СТАЦИОНАРАХ ------
    data_new_year_elg = df_elg.loc[
                                   (df_elg['Дата_госпит'] <= date_start)
                               &   ( (df_elg['Дата_выбытия'].isnull())  | (df_elg['Дата_выбытия'] >= date_start)   
    )]
    
    data_new_year_elg = data_new_year_elg.drop_duplicates('Ключ')
    
    
    # ---- было на дату date_finish в СТАЦИОНАРАХ ------ 
    count_on_date_elg= df.loc[ 
                               (df['Дата_госпит'] <= date_finish)                                           
                           &   ((df['Дата_выбытия'].isnull()) | (df['Дата_выбытия'] >= date_finish))    
                           # &   (df['Категория'] == 'Старше трудосп')].shape[0]
                              ]
    count_on_date_elg = count_on_date_elg.drop_duplicates('Ключ')
    
    return data_new_year_elg.shape[0], count_on_date_elg.shape[0]
```
