```python
# ------------------------ НОРМАЛИЗАЦИЯ ВРЕМЕНИ ---------------------
data_elg['Дата фактического выбытия'] = pd.to_datetime(data_elg['Дата фактического выбытия'], dayfirst=True, errors = 'coerce').dt.normalize()
data_elg['Фактическая дата госпитализации'] = pd.to_datetime(data_elg['Фактическая дата госпитализации'], dayfirst=True, errors = 'coerce').dt.normalize()
data_elg.shape
```

```python
data_elg.kd = np.nan
data_elg.loc[~data_elg['Дата фактического выбытия'].isna(), 'kd'] = data_elg['Дата фактического выбытия'] - data_elg['Фактическая дата госпитализации'] 
data_elg.loc[data_elg['Дата фактического выбытия'].isna(), 'kd'] = pd.to_datetime('today') - data_elg['Фактическая дата госпитализации'] 
data_elg.head(3)
```

```python
# ----- берем только дни ---
data_elg['kd'] = data_elg['kd'].apply ( lambda x: x.days)
```

```python
# ----- удалим менее 0 и боле 2х лет ----------------
data_elg = data_elg.loc[~(data_elg['kd'] < 0)]
data_elg = data_elg.loc[~(data_elg['kd'] > 365 * 2)]
```
